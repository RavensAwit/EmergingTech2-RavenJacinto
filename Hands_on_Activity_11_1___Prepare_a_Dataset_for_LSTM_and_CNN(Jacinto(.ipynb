## **Scale the training set from 0 to 1. Use MinMaxScaler and fit_transform function to do this**

```python
from sklearn.preprocessing import MinMaxScaler

# Exclude non-numeric data in the dataset
numeric_training_set = training_set.select_dtypes(include=['number'])

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(numeric_training_set)
scaled_training_set = pd.DataFrame(scaled_features, columns=numeric_training_set.columns)

# Add the non-numeric data in the dataset
for column in training_set.columns:
    if column not in numeric_training_set.columns:
        scaled_training_set[column] = training_set[column]

print(scaled_training_set.head())
```

Since numerical and non-numerical value exists in data set, it cannot be directly scaled using MinMax. What I did is I separate the number data types from non-numerical data types and then scale it. I turned it back to a dataframe "scaled_training_set". I did a for loop that checks if the current column is not present in the numeric_training_set DataFrame's columns, if a column is found in training_set but not in numeric_training_set, it implies that the column contains non-numeric data. This operation essentially ensures that non-numeric columns are retained in the scaled_training_set DataFrame.

## **LSTM stores long-term memory states. To do this, create a data structure with 60 timesteps and 1 output. Thus, for each element of the training set, we shall have 60 previous training set elements.**

```python
import numpy as np

training_data = scaled_training_set.values

X_train = []
y_train = []

timesteps = 60

# Creating input sequences and corresponding outputs
for i in range(timesteps, len(training_data)):
    X_train.append(training_data[i - timesteps:i, :])
    y_train.append(training_data[i, 0])
```

Converts data to NumPy array to make it more suitable for numerical computations. Stored the input sequences and corresponding output to empty list. For input sequence, it contains 60 timesteps. I created input sequences and corresponding outputs for training the LSTM model. I append an input sequence to the list X_train and the corresponding output to the list y_train.

## **Reshape the X_train for efficient modeling**

```python
# Turning it to array
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshaping
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

Reshaped the data for it to be processed in the model because LSTM models expect input data to have a specific shape. Therefore, reshaping makes it compatible with the input requirements of the LSTM model.
